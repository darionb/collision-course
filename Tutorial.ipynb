# Particle and Evolution Tutorial

```python
from particle import Particle
import evolution as evo
```

## Create a system of particles

> Particle takes in ten possible arguments.
 See [documentation](https://github.com/darionb/collision-course/blob/main/particle.py).

```python
body1 = Particle(velocity=[2,0])
body2 = Particle(position=(5,5),velocity=[0,-1])
system = [body1,body2] 
```

## PlotSystem

Evolution.PlotSystem was designed to work with the class, Particle.

PlotSystem takes an iterable set of particle objects and plots their positions and velocity vectors.

```python
evo.plotSystem(system)
```

```python
p1 = Particle(mass=10,radius=2,position=[0,0],velocity=[3,0]) # create two new particles on the x-axis
p2 = Particle(5,2,[3,0],[-1,0])
system = [p1,p2]

evo.plotSystem(system) #plot system at time = 0
```

## PlayTime
'playTime' uses the function 'evolveOneStep' to discretley advance the position and velocity of particles in the system. PlayTime runs evolveOneStep until the entire system has evolved for the given duration.

> Note: evolveOneStep is able to handle elastic collisions by referencing built-in functions. see [documentation](https://github.com/darionb/collision-course/blob/main/evolution.py).

```python
dt = 3 # in seconds
evo.playTime(system,dt) 
```

playTime does not show us our system.

If we want to see howit has evolved we can use the plotSystem function.

```python
evo.plotSystem(system)
```

## PlayTimePlot
Plot the system. Evolve it. Plot the system again.

If we want to do it all in one go we can use playTimePlot.

```python
p1 = Particle(mass=10,radius=2,position=[0,0],velocity=[3,0]) # reinitializing p1 and p2
p2 = Particle(5,2,[3,0],[-1,0])
system = [p1,p2]
for p in system:
    print (p)
```

```python
dt = 3.5
evo.playTimePlot(system,dt,labels=True) #labels = true enables the start and end indicators beside each particle.
```
